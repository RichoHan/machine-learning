```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use('seaborn-whitegrid')
```

## Experiment
Based on 10000 round of AdaDelta gradient descent.

```python
exp = pd.read_csv('./data/validation_exp.csv', encoding="big5")

sns.set_style("darkgrid")
fig, (ax1, ax2) = plt.subplots(2, figsize=(15, 10))

ax1.plot(exp.loc[:, 'cost'])
ax1.set_ylabel('Cost')

ax2.plot(exp.loc[:, 'score'])
ax2.set_ylabel('Accuracy')

plt.xlabel('Round')
plt.show()
```

```python
exp = pd.read_csv('./data/submission_exp.csv', encoding="big5")

sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(exp.loc[:, 'cost'])
ax.set_ylabel('Cost')

plt.xlabel('Round')
plt.show()
```
