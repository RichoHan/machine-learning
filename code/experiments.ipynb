```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use('seaborn-whitegrid')
```

## Experiment
Based on 20000 iterations of AdaDelta gradient descent without regularization.

```python
exp = pd.read_csv('./data/validation_exp_no_regularization.csv', encoding="big5")

sns.set_style("darkgrid")
fig, (ax1, ax2) = plt.subplots(2, figsize=(15, 10))

ax1.plot(exp.loc[:, 'cost'])
ax1.set_ylabel('Cost')

ax2.plot(exp.loc[:, 'score'])
ax2.set_ylabel('Accuracy')

plt.xlabel('Iteration')
plt.show()
```

```python
exp = pd.read_csv('./data/regularization_validation.csv', encoding="big5")

plt.subplots(figsize=(15, 5))
ax = sns.pointplot(
    x="regularization", y="score",
    data=exp,
    ci=None
)
plt.show()
```

Experiment for the final submission on test data with regularization factor λ=100.

```python
exp = pd.read_csv('./data/submission_exp.csv', encoding="big5")

sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(exp.loc[:, 'cost'])
ax.set_ylabel('Cost')

plt.xlabel('Iteration')
plt.show()
```
