```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.style.use('seaborn-whitegrid')
```

## Experiment
Based on 10000 round of AdaDelta gradient descent.

```python
exp = pd.read_csv('./data/exp.csv', encoding="big5")

sns.set_style("darkgrid")
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(exp.loc[:, 'cost'], label='Cost')
ax.plot(exp.loc[:, 'score'], label='Accuracy')

from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('large')
ax.legend(
    prop = fontP,
    bbox_to_anchor=(1, 1)
)

plt.xlabel('Round')
plt.show()
```
