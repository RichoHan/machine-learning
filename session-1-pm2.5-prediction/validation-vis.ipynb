```python
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")

validation = pd.read_csv('./data/regulation_validation.csv')
processed_validation = None

for index, row in validation.iterrows():
    new_row = pd.DataFrame([
        [int(row['regularization']), row['testing scores'], 'test'],
        [int(row['regularization']), row['training scores'], 'train']
    ], columns=['regularization', 'score', 'type'])
    if processed_validation is None:
        processed_validation = new_row
    else:
        processed_validation = processed_validation.append(new_row, ignore_index=True)

ax = sns.pointplot(
    x="regularization", y="score", hue="type",
    data=processed_validation,
    palette="Set2",
    ci=None
)
sns.plt.show()
```

```python
validation = pd.read_csv('./data/windows_validation.csv')
processed_validation = None

for index, row in validation.iterrows():
    new_row = pd.DataFrame([
        [int(row['window size']), row['testing score'], 'test'],
        [int(row['window size']), row['training score'], 'train']
    ], columns=['window size', 'score', 'type'])
    if processed_validation is None:
        processed_validation = new_row
    else:
        processed_validation = processed_validation.append(new_row, ignore_index=True)

ax = sns.pointplot(
    x="window size", y="score", hue="type",
    data=processed_validation,
    palette="Set2",
    ci=None
)
sns.plt.show()
```

```python
validation = pd.read_csv('./data/cross_validation.csv')
processed_validation = None

for index, row in validation.iterrows():
    new_row = pd.DataFrame([
        [int(row['regularization']), int(row['window size']), row['testing score'], 'test'],
        [int(row['regularization']), int(row['window size']), row['training score'], 'train']
    ], columns=['regularization', 'window size', 'score', 'type'])
    if processed_validation is None:
        processed_validation = new_row
    else:
        processed_validation = processed_validation.append(new_row, ignore_index=True)

ax = sns.pointplot(
    x="regularization", y="score", hue="type",
    data=processed_validation,
    palette="Set2",
    ci=None,
    markers=["o", "x"],
    linestyles=["-", "--"]
)

sns.plt.show()
```

```python
validation = pd.read_csv('./data/features_validation.csv')
processed_validation = None

for index, row in validation.iterrows():
    new_row = pd.DataFrame([
        [row['feature'], int(row['regularization']), int(row['window size']), row['testing score'], 'test'],
        [row['feature'], int(row['regularization']), int(row['window size']), row['training score'], 'train']
    ], columns=['feature', 'regularization', 'window size', 'score', 'type'])
    if processed_validation is None:
        processed_validation = new_row
    else:
        processed_validation = processed_validation.append(new_row, ignore_index=True)

ax = sns.factorplot(
    x="feature", y="score", hue="type",
    kind="bar",
    data=processed_validation,
    palette="Set2",
    ci=95,
    size=6,
    aspect=2
)

sns.plt.show()
```
