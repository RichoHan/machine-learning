```python
import pandas as pd
import seaborn as sns

sns.set_style("darkgrid")

validation = pd.read_csv('./data/linear_validation.csv')
processed_validation = None

for index, row in validation.iterrows():
    new_row = pd.DataFrame([
        [int(row['regularization']), row['testing scores'], 'test'],
        [int(row['regularization']), row['training scores'], 'train']
    ], columns=['regularization', 'score', 'type'])
    if processed_validation is None:
        processed_validation = new_row
    else:
        processed_validation = processed_validation.append(new_row, ignore_index=True)

ax = sns.pointplot(
    x="regularization", y="score", hue="type",
    data=processed_validation,
    palette="Set2"
)
sns.plt.show()
```
